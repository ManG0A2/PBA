```python
import sys

# 성적 데이터 로드 함수
def load_data(students.txt):
    try:
        with open(students.txt, 'r') as file:
            data = [line.strip().split('\t') for line in file]
            return data
    except FileNotFoundError:
        print(f"Error: {students.txt} not found.")
        sys.exit(1)

# 평균 점수 계산 함수
def calculate_average(midterm, final):
    return (float(midterm) + float(final)) / 2

# 학점 계산 함수
def calculate_grade(avg):
    if avg >= 90:
        return 'A'
    elif avg >= 80:
        return 'B'
    elif avg >= 70:
        return 'C'
    elif avg >= 60:
        return 'D'
    else:
        return 'F'

# 성적 데이터 표시 함수
def show(data):
    sorted_data = sorted(data, key=lambda x: calculate_average(x[2], x[3]), reverse=True)
    print("학번\t이름\t중간고사\t기말고사\t평균\t학점")
    for entry in sorted_data:
        avg = calculate_average(entry[2], entry[3])
        grade = calculate_grade(avg)
        print(f"{entry[0]}\t{entry[1]}\t{entry[2]}\t{entry[3]}\t{avg:.1f}\t{grade}")

# 학생 검색 함수
def search(data, student_id):
    for entry in data:
        if entry[0] == student_id:
            avg = calculate_average(entry[2], entry[3])
            grade = calculate_grade(avg)
            print(f"{entry[0]}\t{entry[1]}\t{entry[2]}\t{entry[3]}\t{avg:.1f}\t{grade}")
            return
    print("NO SUCH PERSON.")

# 성적 변경 함수
def changescore(data, student_id):
    for entry in data:
        if entry[0] == student_id:
            exam_type = input("Mid/Final? ").strip().lower()
            if exam_type not in ['mid', 'final']:
                return
            new_score = float(input("Input new score: "))
            if new_score < 0 or new_score > 100:
                return
            if exam_type == 'mid':
                entry[2] = str(int(new_score))
            else:
                entry[3] = str(int(new_score))
            print("Score changed.")
            return
    print("NO SUCH PERSON.")

# 학생 추가 함수
def add(data, student_number, name, midterm, final):
    for entry in data:
        if entry[0] == student_number:
            print("ALREADY EXISTS.")
            return
    data.append([student_number, name, midterm, final])
    print("Student added.")

# 학생 제거 함수
def remove(data, student_id):
    if not data:
        print("List is empty.")
        return
    for entry in data:
        if entry[0] == student_id:
            data.remove(entry)
            print("Student removed.")
            return
    print("NO SUCH PERSON.")

# 종료 시 데이터 저장 함수
def save_data(data):
    save_option = input("Save data?[yes/no] ").strip().lower()
    if save_option == 'yes':
        filename = input("File name: ").strip()
        with open(filename, 'w') as file:
            for entry in sorted(data, key=lambda x: calculate_average(x[2], x[3]), reverse=True):
                file.write('\t'.join(entry) + '\n')

# 메인 함수
def main():
    data = load_data("scores.txt")
    while True:
        command = input("명령어를 입력하세요 (show, search, changescore, add, remove, quit): ").strip().lower()
        if command == "show":
            show(data)
        elif command == "search":
            student_id = input("Student ID: ").strip()
            search(data, student_id)
        elif command == "changescore":
            student_id = input("Student ID: ").strip()
            changescore(data, student_id)
        elif command == "add":
            student_number = input("Student ID: ").strip()
            name = input("Name: ").strip()
            midterm = input("Midterm Score: ").strip()
            final = input("Final Score: ").strip()
            add(data, student_number, name, midterm, final)
        elif command == "remove":
            student_id = input("Student ID: ").strip()
            remove(data, student_id)
        elif command == "quit":
            save_data(data)
            print("$")
            break
        else:
            print("#", end=" ")

if __name__ == "__main__":
    main()
```
